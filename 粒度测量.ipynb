```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('particle_size.tif')
cv2.imshow('origin',img)
blur = cv2.blur(img, (5,5))
# kernel = np.ones((5,5), np.float32)/25
# blur = cv2.filter2D(img, -1, kernel)
###  注意，opencv分离的图像为bgr，而不是rgb
b,g,r = cv2.split(img)

cv2.imshow('blur', img)

def sum_image(image):
    b,g,r = cv2.split(img)
    b_sum = sum(sum(b[i]) for i in range(len(b)))
    g_sum = sum(sum(g[i]) for i in range(len(g)))
    r_sum = sum(sum(r[i]) for i in range(len(r)))
    all_sum = b_sum + g_sum + r_sum
    return all_sum

def sum_image_b(image):
    sum_b = sum(sum(image[i]) for i in range(len(image)))
    return sum_b

blur_b = cv2.blur(b, (5,5))
x=list(range(1,86,2))
y=[0,0]

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
opening = cv2.morphologyEx(blur_b, cv2.MORPH_OPEN, kernel)
fore_sum = sum_image_b(opening)
# print(fore_sum)

for i in range(5,86,2):
#     kernel = shape_to_circle(i)
#     print(kernel)
#     print(kernel.shape)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(i,i))
    opening = cv2.morphologyEx(blur_b, cv2.MORPH_OPEN, kernel)
    if(i%5 == 0):
        cv2.imshow(str(i)+' open',opening)
    next_sum = sum_image_b(opening)
#     print(next_sum)
    difference_next_fore = fore_sum - next_sum
    y.append(difference_next_fore)
    fore_sum = next_sum

print(x)
print(y)
plt.figure(num=1, figsize=(10,6))#定义一个图像窗口，编号为1，大小(10,6)
plt.plot(x,y,'o-')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
